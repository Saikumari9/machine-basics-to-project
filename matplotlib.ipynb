```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(1,11)
y=x**2
print(x,y)
```

```python
plt.plot(x,y)
plt.title('Square Funtion Line plot')
plt.xlabel('N value')
plt.ylabel('Square of N')
plt.show()
```
